Q.1)   What is Anomaly Detection? Explain its types (point, contextual, and collective anomalies) with examples. 
Ans:- Anomaly Detection is the process of identifying patterns in data that do not conform to expected behavior. These unusual patterns are called anomalies or outliers. Detecting anomalies is crucial in many domains such as fraud detection, network security, fault detection in systems, and medical diagnosis.

Types of Anomalies
1. Point Anomalies
Definition: A single data instance is anomalous compared to the rest of the dataset.
Example:
- In a credit card transaction dataset, a purchase of $10,000 when most transactions are between $10–$500.
- In sensor readings, a sudden spike in temperature from 30°C to 100°C.
Use Case: Fraud detection, sensor malfunction alerts.

2. Contextual Anomalies
Definition: A data instance is anomalous in a specific context but may be normal in another.
Example:
- A temperature of 30°C is normal in summer but anomalous in winter.
- A website receiving 10,000 hits in a day is normal during a sale but abnormal on a regular day.
Use Case: Time-series analysis, seasonal data monitoring.

3. Collective Anomalies
Definition: A collection of related data instances is anomalous, even if individual points are not.
Example:
- In network traffic, a sudden burst of continuous requests from a single IP address (possible DDoS attack).
- In stock market data, a sequence of small declines forming a suspicious downward trend.
Use Case: Cybersecurity, system fault detection, financial market monitoring.


Q.2) Compare Isolation Forest, DBSCAN, and Local Outlier Factor in terms of their approach and suitable use cases.
Ans:-
Isolation Forest works by randomly partitioning the dataset and isolating points. Since anomalies are easier to isolate, they require fewer splits compared to normal points. This makes the algorithm highly efficient for large, high-dimensional datasets. It is particularly suitable for fraud detection, cybersecurity, and other large-scale anomaly detection tasks where scalability is important.

DBSCAN (Density-Based Spatial Clustering of Applications with Noise) is a clustering algorithm that groups points into dense clusters and labels points in sparse regions as noise or anomalies. It does not require specifying the number of clusters and can discover clusters of arbitrary shapes. DBSCAN is best suited for spatial data, sensor networks, and applications where anomalies naturally appear as isolated points outside dense clusters.

Local Outlier Factor (LOF) takes a localized approach by comparing the density of each point to the density of its neighbors. Points with significantly lower density than their neighbors are flagged as anomalies. This makes LOF effective in datasets with varying densities, where anomalies are relative to their local context. It is commonly used in network intrusion detection, fraud detection with localized variations, and time-series data where local deviations matter.

In summary, Isolation Forest is ideal for large-scale, high-dimensional problems, DBSCAN is useful for density-based anomaly detection in spatial data, and LOF is most effective when anomalies are defined relative to local neighborhoods.


Q.3) What are the key components of a Time Series? Explain each with one example.
Ans:- A time series is a sequence of data points collected or recorded at successive points in time, usually at equal intervals. It is widely used in forecasting, anomaly detection, and trend analysis. The key components of a time series are trend, seasonality, cyclic patterns, and irregular (random) variations. Let’s go through each one with an example:

1. Trend  
The trend represents the long-term movement or direction in the data over time. It shows whether the values are increasing, decreasing, or remaining stable.
Example: The steady rise in global average temperatures over decades due to climate change reflects a clear upward trend.

2. Seasonality  
Seasonality refers to regular and predictable patterns that repeat over a fixed period, such as daily, monthly, or yearly.
Example: Retail sales often spike every December due to holiday shopping, showing a seasonal effect.

3. Cyclic Patterns  
Cyclic variations are long-term oscillations in the data that occur over periods longer than a year. Unlike seasonality, cycles do not have a fixed or predictable length.
Example: Economic data showing periods of boom and recession illustrates cyclic behavior.

4. Irregular (Random) Variations  
These are unpredictable fluctuations caused by unexpected events or random noise. They do not follow a pattern and are often short-lived.
Example: A sudden drop in airline ticket sales due to a natural disaster or pandemic is an irregular variation.


Q.4) Define Stationary in time series. How can you test and transform a non-stationary series into a stationary one? 
Ans:- In time series analysis, a series is said to be stationary if its statistical properties such as mean, variance, and autocorrelation remain constant over time. In other words, the behavior of the series does not depend on when you observe it. Stationarity is important because many forecasting models (like ARIMA) assume the data is stationary to make reliable predictions.

How to Test Stationarity:
1. Visual Inspection: Plot the time series and check if the mean and variance appear constant over time.
Example: A series with a clear upward trend is likely non-stationary.

2. Summary Statistics: Divide the series into different time intervals and compare the mean and variance across intervals.
Example: If the average sales in the first half of the data is 200 units and in the second half is 500 units, the series is non-stationary.

3. Statistical Tests: Use tests like the Augmented Dickey-Fuller (ADF) test or KPSS test.
Example: If the ADF test p-value > 0.05, we fail to reject the null hypothesis of non-stationarity.

How to Transform Non-Stationary Series into Stationary:
1. Differencing: Subtract the current value from the previous value to remove trends.
Example: If monthly sales show a steady upward trend, applying first-order differencing (sales[t] – sales[t-1]) can stabilize the mean.

2. Detrending: Fit a regression line to the trend and subtract it from the series.
Example: Removing a linear upward trend in temperature data.

3. Seasonal Adjustment: Remove seasonal effects by subtracting seasonal averages.
Example: Adjusting retail sales data by removing holiday season spikes.

4. Transformation (Log/Box-Cox): Apply mathematical transformations to stabilize variance.
Example: Taking the logarithm of stock prices to reduce variance fluctuations.

So, a stationary series is one whose statistical properties remain constant over time. You can test it using visual inspection, statistical tests like ADF, and then transform it using differencing, detrending, seasonal adjustment, or variance-stabilizing transformations.


Q.5)Differentiate between AR, MA, ARIMA, SARIMA, and SARIMAX models in terms of structure and application. 
Ans:- AR, MA, ARIMA, SARIMA, and SARIMAX models in terms of their structure and applications, explained one by one:

1. Autoregressive (AR) Model  
The AR model predicts future values based on a linear combination of past values. It assumes that past observations have a direct influence on the current value.
Example: Forecasting stock prices where today’s price depends on the last few days’ prices.
Application: Suitable for time series with strong autocorrelation.

2. Moving Average (MA) Model  
The MA model predicts future values based on past forecast errors (residuals). It assumes that the current value is influenced by random shocks from previous time steps.
Example: Modeling daily temperature fluctuations where today’s deviation depends on past random disturbances.
Application: Useful when noise or shocks significantly affect the series.

3. Autoregressive Integrated Moving Average (ARIMA) Model  
ARIMA combines AR and MA components with differencing (the “I” part) to make a non-stationary series stationary. It captures both autocorrelation and random shocks while handling trends.
Example: Forecasting monthly sales data after removing trend and seasonality.
Application: Widely used for general-purpose forecasting in economics, finance, and business.

4. Seasonal ARIMA (SARIMA) Model  
SARIMA extends ARIMA by explicitly modeling seasonality. It includes seasonal autoregressive, seasonal differencing, and seasonal moving average terms.
Example: Forecasting electricity demand that shows daily and yearly seasonal patterns.
Application: Best for time series with strong seasonal cycles, such as retail sales or weather data.

5. Seasonal ARIMA with Exogenous Variables (SARIMAX) Model  
SARIMAX further extends SARIMA by incorporating external (exogenous) variables that may influence the series. These variables provide additional explanatory power.
Example: Forecasting airline passenger demand using SARIMAX with exogenous variables like fuel prices or holidays.
Application: Useful when external factors strongly impact the time series, making forecasts more accurate.


```python
#Q.6) Load a time series dataset (e.g., AirPassengers), plot the original series, and decompose it into trend, seasonality, and residual components
#Ans:-
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose

# Step 1: Load the dataset
data = pd.read_csv("AirPassengers (1).csv", parse_dates=['Month'], index_col='Month')

# Step 2: Plot the original time series
plt.figure(figsize=(10,5))
plt.plot(data, label='AirPassengers')
plt.title("AirPassengers Time Series")
plt.xlabel("Year")
plt.ylabel("Number of Passengers")
plt.legend()
plt.show()

# Step 3: Decompose the time series
decomposition = seasonal_decompose(data['#Passengers'], model='multiplicative', period=12)

# Step 4: Plot the decomposition
fig = decomposition.plot()
fig.set_size_inches(12, 8)
plt.show()
```

```python
#Q.7) Apply Isolation Forest on a numerical dataset to detect anomalies. Visualize the anomalies on a 2D scatter plot.
#Ans:-
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest

# Step 1: Generate synthetic numerical dataset
rng = np.random.RandomState(42)
X = 0.3 * rng.randn(100, 2)  # cluster 1
X = np.r_[X + 2, X - 2]      # cluster 2
X_outliers = rng.uniform(low=-6, high=6, size=(20, 2))  # anomalies
X = np.r_[X, X_outliers]

# Step 2: Fit Isolation Forest
clf = IsolationForest(contamination=0.1, random_state=42)
clf.fit(X)
y_pred = clf.predict(X)

# Step 3: Visualize anomalies
plt.figure(figsize=(8,6))
plt.scatter(X[:, 0], X[:, 1], c='blue', label='Normal Points')
plt.scatter(X[y_pred == -1, 0], X[y_pred == -1, 1], 
            c='red', marker='x', label='Anomalies')
plt.title("Isolation Forest Anomaly Detection")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.show()
```

```python
#Q.8) Train a SARIMA model on the monthly airline passengers dataset. Forecast the next 12 months and visualize the results.
#Ans:-
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.sarimax import SARIMAX

# Step 1: Load the dataset
data = pd.read_csv("AirPassengers (1).csv", parse_dates=['Month'], index_col='Month')
ts = data['#Passengers']

# Step 2: Fit SARIMA model
# Order = (p,d,q), Seasonal_order = (P,D,Q,s)
# Here we use (1,1,1) and seasonal (1,1,1,12) as a common choice for monthly seasonal data
model = SARIMAX(ts, order=(1,1,1), seasonal_order=(1,1,1,12))
results = model.fit(disp=False)

# Step 3: Forecast next 12 months
forecast = results.get_forecast(steps=12)
forecast_mean = forecast.predicted_mean
forecast_ci = forecast.conf_int()

# Step 4: Visualization
plt.figure(figsize=(12,6))
plt.plot(ts, label='Observed')
plt.plot(forecast_mean, label='Forecast', color='red')
plt.fill_between(forecast_ci.index,
                 forecast_ci.iloc[:,0],
                 forecast_ci.iloc[:,1], color='pink', alpha=0.3)
plt.title("SARIMA Forecast of Airline Passengers")
plt.xlabel("Year")
plt.ylabel("Number of Passengers")
plt.legend()
plt.show()
```

```python
#Q.9)Apply Local Outlier Factor (LOF) on any numerical dataset to detect anomalies and visualize them using matplotlib.
#Ans:-
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import LocalOutlierFactor

# Step 1: Generate synthetic dataset
rng = np.random.RandomState(42)
X = 0.3 * rng.randn(100, 2)   # cluster 1
X = np.r_[X + 2, X - 2]       # cluster 2
X_outliers = rng.uniform(low=-6, high=6, size=(20, 2))  # anomalies
X = np.r_[X, X_outliers]

# Step 2: Apply Local Outlier Factor
lof = LocalOutlierFactor(n_neighbors=20, contamination=0.1)
y_pred = lof.fit_predict(X)

# Step 3: Visualization
plt.figure(figsize=(8,6))
plt.scatter(X[y_pred == 1, 0], X[y_pred == 1, 1], 
            c='blue', label='Normal Points')
plt.scatter(X[y_pred == -1, 0], X[y_pred == -1, 1], 
            c='red', marker='x', label='Anomalies')
plt.title("Local Outlier Factor (LOF) Anomaly Detection")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.show()
```

```python
#Q.10)You are working as a data scientist for a power grid monitoring company. 
#Your goal is to forecast energy demand and also detect abnormal spikes or drops in real-time consumption data collected every 15 minutes. The dataset includes features like timestamp, region, weather conditions, and energy usage. 
#Explain your real-time data science workflow: 
#● How would you detect anomalies in this streaming data (Isolation Forest / LOF / DBSCAN)? 
#● Which time series model would you use for short-term forecasting (ARIMA / SARIMA / SARIMAX)? 
#● How would you validate and monitor the performance over time? 
#● How would this solution help business decisions or operations? 
#Ans:-
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest
from statsmodels.tsa.statespace.sarimax import SARIMAX

# Simulated streaming dataset
np.random.seed(42)
timestamps = pd.date_range("2023-01-01", periods=200, freq="15min")
energy_usage = 50 + np.sin(np.linspace(0, 20, 200)) * 10 + np.random.normal(0, 2, 200)
# Inject anomalies
energy_usage[50] = 120   # spike
energy_usage[120] = 10   # drop

df = pd.DataFrame({"timestamp": timestamps, "energy_usage": energy_usage})
df.set_index("timestamp", inplace=True)

# --- Anomaly Detection with Isolation Forest ---
clf = IsolationForest(contamination=0.05, random_state=42)
df["anomaly"] = clf.fit_predict(df[["energy_usage"]])

# --- Forecasting with SARIMAX ---
model = SARIMAX(df["energy_usage"], order=(1,1,1), seasonal_order=(1,1,1,96)) # 96 = 24h/15min
results = model.fit(disp=False)
forecast = results.get_forecast(steps=48)  # next 12 hours
forecast_mean = forecast.predicted_mean
forecast_ci = forecast.conf_int()

# --- Visualization ---
plt.figure(figsize=(12,6))
plt.plot(df.index, df["energy_usage"], label="Observed")
plt.scatter(df.index[df["anomaly"]==-1], df["energy_usage"][df["anomaly"]==-1],
            color="red", marker="x", label="Anomalies")
plt.plot(forecast_mean.index, forecast_mean, color="green", label="Forecast")
plt.fill_between(forecast_ci.index, forecast_ci.iloc[:,0], forecast_ci.iloc[:,1],
                 color="lightgreen", alpha=0.3)
plt.title("Energy Demand Forecast & Anomaly Detection")
plt.xlabel("Time")
plt.ylabel("Energy Usage (MW)")
plt.legend()
plt.show()
```
