Que.1)  What is Simple Linear Regression (SLR)? Explain its purpose. 
Ans:- Simple Linear Regression (SLR)is a statistical technique used to model the relationship between two variables:
- Independent variable (X): The predictor or input.
- Dependent variable (Y): The outcome or response.

It assumes that the relationship between X and Y can be represented by a straight line:
𝑌=𝛽0+𝛽1𝑋+𝜖
where, 
𝛽0: Intercept (value of Y when X = 0)
𝛽1: Slope (change in Y for a one-unit change in X)
𝜖: Error term (captures variability not explained by X)

Purpose of SLR:-
The main goals of Simple Linear Regression are:
1. Prediction: Estimate the value of Y based on a given X.
Example: Predicting a student’s exam score based on study hours.
2. Understanding relationships: Quantify how strongly X influences Y.
Example: Measuring how much sales increase for every extra dollar spent on advertising.
3. Trend identification: Detect whether the relationship is positive, negative, or negligible.
Example: Identifying if higher temperatures lead to increased ice cream sales.
4. Decision-making: Provide insights for business, science, and policy by showing how one factor impacts another.

In short, SLR is the simplest form of regression analysis, acting as a foundation for more complex models. It’s widely used because it’s easy to interpret and provides clear insights into cause-and-effect relationships.

Que.2) What are the key assumptions of Simple Linear Regression? 
Ans:- Key Assumptions of Simple Linear Regression (SLR):
For SLR to provide valid and reliable results, several assumptions must hold true:
1. Linearity:-
- The relationship between the independent variable (X) and the dependent variable (Y) is linear.
- Meaning: A straight line can adequately describe how Y changes with X.

2. Independence of Errors:-
- The residuals (errors) are independent of each other.
- Example: In time series data, this means no autocorrelation between errors.

3. Homoscedasticity (Constant Variance of Errors):-
- The variance of residuals is constant across all values of X.
- If variance changes (heteroscedasticity), predictions become less reliable.

4. Normality of Errors
- The residuals should be approximately normally distributed.
- Important for hypothesis testing and confidence intervals.

5. No Perfect Multicollinearity
- Since SLR uses only one predictor, this assumption is trivial here.
- But in multiple regression, predictors must not be perfectly correlated.

6. X is Measured Without Error
- The independent variable is assumed to be measured accurately.
- Measurement errors in X can bias the slope estimate.

Que.3) Write the mathematical equation for a simple linear regression model and explain each term. 
Ans:- The general form of a Simple Linear Regression (SLR) model is:
𝑌=𝛽0+𝛽1𝑋+𝜖

Explanation of Each Term:-
1. 𝑌(Dependent Variable / Response):  The outcome we are trying to predict or explain.
Example: House price.
2.𝑋(Independent Variable / Predictor):  he input variable used to predict 𝑌.
Example: Square footage of the house.
3. 𝛽0(Intercept):  The value of𝑌 when 𝑋=0. It represents the baseline level of the dependent variable.
Example: The predicted house price when square footage is zero (often just a mathematical anchor).
4. 𝛽1(Slope Coefficient):  The change in 𝑌 for a one-unit increase in 𝑋. It quantifies the strength and direction of the relationship.
Example: If 𝛽1=200, then each additional square foot increases the house price by $200.
5. 𝜖 (Error Term / Residual):  Captures the variation in 𝑌 that cannot be explained by 𝑋. It accounts for randomness, measurement error, or other factors not included in the model.

Que.5)What is the method of least squares in linear regression? 
Ans:- The method of least squares is the most common technique used to estimate the parameters (𝛽0 and 𝛽1) in a simple linear regression model.
How It Works:-
- The regression line is meant to fit the data points as closely as possible.
- For each observation, the residual is the difference between the actual value (𝑌𝑖) and the predicted value (𝑌^𝑖):
Residual𝑖=𝑌𝑖−𝑌^𝑖
- The least squares method minimizes the sum of squared residuals (SSR):
𝑆𝑆𝑅=∑𝑖=1𝑛(𝑌𝑖−(𝛽0+𝛽1𝑋𝑖))2
By minimizing this sum, we find the line that best represents the relationship between 𝑋 and 𝑌.

Purpose:-
1. Best Fit Line: Ensures the regression line is as close as possible to all data points collectively.
2. Unbiased Estimates: Provides estimates of slope and intercept that are statistically optimal under the regression assumptions.
3. Predictive Accuracy: Improves the reliability of predictions for new values of 𝑋.

Que.6) What is Logistic Regression? How does it differ from Linear Regression? 
Ans:- Logistic Regression is used for classification problems (predicting categorical outcomes like Yes/No), while Linear Regression is used for regression problems (predicting continuous values like prices or scores). The key difference lies in the type of dependent variable and the mathematical function used to model it.

Logistic Regression
Purpose: Predicts the probability of a categorical outcome (often binary: 0 or 1).
Equation:
𝑃(𝑌=1∣𝑋)=11+𝑒−(𝛽0+𝛽1𝑋)
- Uses the logistic (sigmoid) function to map predictions into the range [0,1].
- Output: Probability values, which are then classified into categories (e.g., spam vs. not spam).
- Applications:
Medical diagnosis (disease vs. no disease)
Email filtering (spam vs. not spam)
Customer churn prediction (leave vs. stay)

Linear Regression
Purpose: Predicts continuous numerical outcomes.
Equation:
𝑌=𝛽0+𝛽1𝑋+𝜖
- Produces values on a continuous scale.
- Output: Direct numerical predictions (e.g., house price, salary, temperature).
- Applications:
Forecasting sales based on advertising spend
Predicting exam scores from study hours
Estimating house prices from square footage

Que.7) Name and briefly describe three common evaluation metrics for regression models.
Ans:- Three Common Evaluation Metrics for Regression Models
When assessing how well a regression model performs, analysts often rely on these metrics:
Mean Absolute Error (MAE)
Formula:
𝑀𝐴𝐸=1𝑛∑𝑖=1𝑛∣𝑌𝑖−𝑌^𝑖∣
Meaning: Average of the absolute differences between actual values (𝑌𝑖) and predicted values (𝑌^𝑖).
Interpretation: Easy to understand since it measures error in the same units as the data. Lower MAE = better fit.

Mean Squared Error (MSE)
Formula:
𝑀𝑆𝐸=1𝑛∑𝑖=1𝑛(𝑌𝑖−𝑌^𝑖)2
Meaning: Average of squared differences between actual and predicted values.
Interpretation: Penalizes larger errors more heavily due to squaring. Useful when large deviations are especially undesirable.

R-squared (𝑅2) – Coefficient of Determination
Formula:
𝑅2=1−(𝑆𝑆𝑅/𝑆𝑆𝑇)
where SSR = sum of squared residuals, SST = total sum of squares.
Meaning: Proportion of variance in the dependent variable explained by the model.
Interpretation: Ranges from 0 to 1. Higher values indicate better explanatory power. For example, 𝑅2=0.85 means 85% of the variation in Y is explained by X.

Que.8) : What is the purpose of the R-squared metric in regression analysis?
Ans:- R-squared (𝑅2), also called the coefficient of determination, measures how well a regression model explains the variability of the dependent variable.
Definition:
𝑅2=1−(𝑆𝑆𝑅/𝑆𝑆𝑇)
SSR (Sum of Squared Residuals): Variation not explained by the model.
SST (Total Sum of Squares): Total variation in the data.

Interpretation:
𝑅2=0: The model explains none of the variability.
𝑅2=1: The model explains all the variability perfectly.
Values between 0 and 1 indicate the proportion of variance explained.

Purpose
- Goodness of Fit: Shows how well the regression line fits the data.
- Model Evaluation: Helps compare different models — higher 𝑅2 generally means better explanatory power.
- Decision Support: Indicates whether the independent variable(s) meaningfully explain changes in the dependent variable.


```python
#Que.9)  Write Python code to fit a simple linear regression model using scikit-learn and print the slope and intercept. 
#Ans:-
import numpy as np
from sklearn.linear_model import LinearRegression

X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)   # Reshape for sklearn
y = np.array([1, 2, 1.3, 3.75, 2.5])

# Create and fit the model
model = LinearRegression()
model.fit(X, y)

# Print slope (coefficient) and intercept
print("Slope (β1):", model.coef_[0])
print("Intercept (β0):", model.intercept_)
```
