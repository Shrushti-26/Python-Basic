Que2)Explain the difference between mutable and immutable data types with examples?
Ans:-i) Mutable: Can be changed after creation. 
Example: list = [1,2,3] → list[0] = 10, modifies it.
ii)Immutable: Cannot be changed after creation.
Example: tuple = (1,2,3) → elements cannot be reassigned.

ii) Tuples = Tuples are immutable(Cannot be changed after creation). "()" symbol is used to denote tuples. Its performance is high due to immutability. Use case = Fixed data.
