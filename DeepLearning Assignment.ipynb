Que.1)What is Deep Learning? Briefly describe how it evolved and how it differs from traditional machine learning. 
Ans:- Deep Learning is a subset of machine learning that uses artificial neural networks with many layers (deep architectures) to automatically learn complex patterns and representations from large amounts of data. It mimics the way the human brain processes information, enabling breakthroughs in areas like computer vision, natural language processing, and speech recognition.

Evolution of Deep Learning
1950s–1980s: Early neural networks (like the perceptron) were introduced, but limited by computational power and small datasets.
1990s: Backpropagation improved training of multi-layer networks, but deep networks still struggled with vanishing gradients and overfitting.
2000s: Advances in algorithms (e.g., ReLU activation, dropout), availability of big data, and GPU acceleration revived interest.
2010s onward: Breakthroughs like AlexNet (2012) in image recognition demonstrated the power of deep learning, leading to widespread adoption in industry and research.

Que.2) Explain the basic architecture and functioning of a Perceptron. What are its limitations?
Ans:- A Perceptron is the simplest type of artificial neural network model, introduced by Frank Rosenblatt in 1958. It is essentially a binary classifier that decides whether input belongs to one class or another.

Components:
1. Inputs (𝑥1,𝑥2,...,𝑥𝑛) – Features of the dataset.
2. Weights (𝑤1,𝑤2,...,𝑤𝑛) – Each input is multiplied by a weight that represents its importance.
3. Summation Function – Computes the weighted sum:
𝑧=∑𝑤𝑖𝑥𝑖+𝑏
where 𝑏 is the bias term.
4. Activation Function – Applies a threshold (step function). If 𝑧≥0, output = 1; else output = 0.

Working:
- Inputs are fed into the perceptron.
- Each input is multiplied by its weight.
- The weighted sum plus bias is calculated.
- The activation function decides the final output (class label).

Limitations of a Perceptron
1. Linear Separability:  
Can only solve problems where data is linearly separable (e.g., AND, OR). It fails on non-linear problems like XOR.

2. Limited Learning Capacity:  
A single-layer perceptron cannot capture complex patterns.

3. No Hidden Layers:  
Without hidden layers, it cannot perform feature extraction or hierarchical learning.

4. Rigid Activation Function:  
Uses a step function, which is non-differentiable, making gradient-based optimization impossible.

Que.3): Describe the purpose of activation function in neural networks. Compare Sigmoid, ReLU, and Tanh functions.
Ans:- The activation function introduces non-linearity into the network. Without it, a neural network would just be a linear model (like a perceptron), unable to capture complex relationships.

Key purposes:
- Allows the network to learn non-linear decision boundaries.
- Controls how signals flow through the network.
- Helps in gradient-based optimization by shaping the output range.
- Enables deep networks to approximate complex functions.

Sigmoid: Best for probabilistic interpretation, but rarely used in hidden layers today.
Tanh: Better than sigmoid since it’s zero-centered, but still limited for deep networks.
ReLU: Default choice in most modern deep learning models due to efficiency and performance.

Que.5) What is the role of optimizers in neural networks? Compare Gradient Descent, Adam, and RMSprop. 
Ans:-Optimizers are algorithms used to adjust the weights and biases of a neural network to minimize the cost (loss) function. They guide the training process by deciding how much and in what direction parameters should change after each iteration.
Key Role:
- Speed up convergence.
- Prevent getting stuck in local minima.
- Balance learning rate for stable training.
- Improve generalization of the model.

Comparison of Gradient Descent, Adam, and RMSprop:-
Gradient Descent: Basic but inefficient for deep networks.
RMSprop: Better for recurrent networks and non-stationary problems.
Adam: Most popular choice today due to efficiency and adaptability.

```python
#Que.6) Write a Python program to implement a single-layer perceptron from scratch using NumPy to solve the logical AND gate.
#Ans:-
import numpy as np

# Step 1: Define input and output for AND gate
# Inputs: (x1, x2), Outputs: AND(x1, x2)
X = np.array([[0,0],
              [0,1],
              [1,0],
              [1,1]])

y = np.array([0, 0, 0, 1])   # AND truth table

# Step 2: Initialize weights and bias
weights = np.zeros(X.shape[1])   # two inputs → two weights
bias = 0
learning_rate = 0.1
epochs = 20

# Step 3: Activation function (Step function)
def activation(z):
    return 1 if z >= 0 else 0

# Step 4: Training loop
for epoch in range(epochs):
    for i in range(len(X)):
        z = np.dot(X[i], weights) + bias
        y_pred = activation(z)
        
        # Error = target - prediction
        error = y[i] - y_pred
        
        # Update rule (Perceptron Learning Rule)
        weights += learning_rate * error * X[i]
        bias += learning_rate * error

    print(f"Epoch {epoch+1}: Weights={weights}, Bias={bias}")

# Step 5: Testing
print("\nTesting Perceptron on AND gate:")
for i in range(len(X)):
    z = np.dot(X[i], weights) + bias
    y_pred = activation(z)
    print(f"Input: {X[i]} → Output: {y_pred}")
```

```python
#Que.7)Implement and visualize Sigmoid, ReLU, and Tanh activation functions using Matplotlib. 
#Ans:-
import numpy as np
import matplotlib.pyplot as plt

# Define activation functions
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def relu(x):
    return np.maximum(0, x)

def tanh(x):
    return np.tanh(x)

# Generate input values
x = np.linspace(-10, 10, 200)

# Compute outputs
y_sigmoid = sigmoid(x)
y_relu = relu(x)
y_tanh = tanh(x)

# Plot activation functions
plt.figure(figsize=(12, 8))

# Sigmoid
plt.subplot(3, 1, 1)
plt.plot(x, y_sigmoid, color='blue')
plt.title("Sigmoid Activation Function")
plt.grid(True)

# ReLU
plt.subplot(3, 1, 2)
plt.plot(x, y_relu, color='green')
plt.title("ReLU Activation Function")
plt.grid(True)

# Tanh
plt.subplot(3, 1, 3)
plt.plot(x, y_tanh, color='red')
plt.title("Tanh Activation Function")
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
#Que.8)Use Keras to build and train a simple multilayer neural network on the MNIST digits dataset. Print the training accuracy.
#Ans:-
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.datasets import mnist
from tensorflow.keras.utils import to_categorical

# Step 1: Load MNIST dataset
(X_train, y_train), (X_test, y_test) = mnist.load_data()

# Step 2: Normalize input data
X_train = X_train / 255.0
X_test = X_test / 255.0

# Step 3: One-hot encode labels
y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)

# Step 4: Build a simple multilayer neural network
model = Sequential([
    Flatten(input_shape=(28, 28)),       # Flatten 28x28 images into 784 vector
    Dense(128, activation='relu'),       # Hidden layer
    Dense(64, activation='relu'),        # Another hidden layer
    Dense(10, activation='softmax')      # Output layer (10 classes)
])

# Step 5: Compile the model
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# Step 6: Train the model
history = model.fit(X_train, y_train, epochs=5, batch_size=32, verbose=1)

# Step 7: Print training accuracy
train_acc = history.history['accuracy'][-1]
print(f"Final Training Accuracy: {train_acc:.4f}")
```

```python
#Que.9)Visualize the loss and accuracy curves for a neural network model trained on the Fashion MNIST dataset. Interpret the training behavior.
#Ans:-
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, Dropout
from tensorflow.keras.datasets import fashion_mnist
from tensorflow.keras.utils import to_categorical

# Step 1: Load Fashion MNIST dataset
(X_train, y_train), (X_test, y_test) = fashion_mnist.load_data()

# Step 2: Normalize input data
X_train = X_train / 255.0
X_test = X_test / 255.0

# Step 3: One-hot encode labels
y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)

# Step 4: Build a simple neural network
model = Sequential([
    Flatten(input_shape=(28, 28)),
    Dense(128, activation='relu'),
    Dropout(0.3),                # Regularization to reduce overfitting
    Dense(64, activation='relu'),
    Dense(10, activation='softmax')
])

# Step 5: Compile the model
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# Step 6: Train the model
history = model.fit(X_train, y_train,
                    epochs=10,
                    batch_size=32,
                    validation_data=(X_test, y_test),
                    verbose=1)

# Step 7: Plot Loss and Accuracy curves
plt.figure(figsize=(12,5))

# Loss curve
plt.subplot(1,2,1)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title("Loss Curve")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()
plt.grid(True)

# Accuracy curve
plt.subplot(1,2,2)
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title("Accuracy Curve")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
#Que.10)You are working on a project for a bank that wants to automatically detect fraudulent transactions. The dataset is large, imbalanced, and contains structured features like transaction amount, merchant ID, and customer location. The goal is to classify each transaction as fraudulent or legitimate. 
#Explain your real-time data science workflow: 
#● How would you design a deep learning model (perceptron or multilayer NN)? 
#● Which activation function and loss function would you use, and why? 
#● How would you train and evaluate the model, considering class imbalance? 
#● Which optimizer would be suitable, and how would you prevent overfitting?
#Ans:-
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix

# -----------------------------
# Step 1: Simulate structured dataset (replace with real bank data)
# -----------------------------
np.random.seed(42)
n_samples = 10000
X = np.random.rand(n_samples, 5)  # features: amount, merchant_id, location, etc.
y = np.random.choice([0,1], size=n_samples, p=[0.95, 0.05])  # imbalanced: 5% fraud

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y)

# -----------------------------
# Step 2: Build multilayer NN
# -----------------------------
model = Sequential([
    Dense(64, activation='relu', input_shape=(X_train.shape[1],)),
    Dropout(0.3),
    Dense(32, activation='relu'),
    Dropout(0.3),
    Dense(1, activation='sigmoid')  # binary output
])

# -----------------------------
# Step 3: Compile with loss & optimizer
# -----------------------------
# Binary crossentropy is standard for fraud detection
# Class weights handle imbalance (fraud class weighted higher)
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

# Compute class weights
from sklearn.utils.class_weight import compute_class_weight
class_weights = compute_class_weight(class_weight='balanced',
                                     classes=np.unique(y_train),
                                     y=y_train)
class_weights_dict = {0: class_weights[0], 1: class_weights[1]}
print("Class Weights:", class_weights_dict)

# -----------------------------
# Step 4: Train with class weights
# -----------------------------
history = model.fit(X_train, y_train,
                    epochs=10,
                    batch_size=32,
                    validation_data=(X_test, y_test),
                    class_weight=class_weights_dict,
                    verbose=1)

# -----------------------------
# Step 5: Evaluate
# -----------------------------
y_pred = (model.predict(X_test) > 0.5).astype(int)
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```
