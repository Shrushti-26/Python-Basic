```python
#Practical Questions:-

#Que1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.
#Ans:-
def sum_even_numbers():
    numbers = input("Enter numbers separated by spaces: ")
    num_list = [int(x) for x in numbers.split()]
    total = sum(num for num in num_list if num % 2 == 0)
    return total

result = sum_even_numbers()
print("Sum of even numbers:", result)
```

```python
#Que2. Create a Python function that accepts a string and returns the reverse of that string.
#Ans:- 
def reverse_string(text):
    """Return the reverse of the given string."""
    return text[::-1]   # slicing with step -1 reverses the string

sample = "Shrushti"
result = reverse_string(sample)
print("Reversed string:", result)
```

```python
#Que4. Write a Python function that checks if a given number is prime or not from 1 to 200.
#Ans:-
def is_prime(n):
    """Check if a number is prime (valid for 1 to 200)."""
    if n <= 1:
        return False   # 0 and 1 are not prime
    for i in range(2, int(n**0.5) + 1):  # check divisibility up to sqrt(n)
        if n % i == 0:
            return False
    return True

for num in range(1, 21):  
    print(num, "is prime?" , is_prime(num))
```

```python
#Que5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.
#Ans:-
def fibonacci(n_terms):
    """Generate Fibonacci sequence up to n_terms using yield."""
    a, b = 0, 1
    for _ in range(n_terms):
        yield a
        a, b = b, a + b

for num in fibonacci(10): 
    print(num, end=" ")
```

```python
#Que6. Write a generator function in Python that yields the powers of 2 up to a given exponent.
#Ans:-
def powers_of_two(n):
    """Yield powers of 2 up to the given exponent n."""
    for i in range(n + 1):   # include n
        yield 2 ** i

for value in powers_of_two(5):
    print(value, end=" ")
```

```python
#Que7. Implement a generator function that reads a file line by line and yields each line as a string.
#Ans:-
def read_file_lines(filename):
    """Generator that yields lines from a file one by one."""
    try:
        with open(filename, "r", encoding="utf-8") as file:
            for line in file:
                yield line.rstrip("\n")  # remove newline character
    except FileNotFoundError:
        print(f"Error: File '{filename}' not found.")
    except Exception as e:
        print(f"Error while reading file: {e}")

# Example usage:
for line in read_file_lines("sample.txt"):
    print(line)


```

```python
#Que9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.
#Ans:-
def c_to_f(celsius):
    return (celsius * 9/5) + 32
celsius_temps = [0, 20, 37, 100]
fahrenheit_temps = list(map(c_to_f, celsius_temps))

print("Celsius:", celsius_temps)
print("Fahrenheit:", fahrenheit_temps)
```

```python
#Que10. Create a Python program that uses `filter()` to remove all the vowels from a given string.
#Ans:-
def remove_vowels(text):
    vowels = "aeiouAEIOU"
    # Using filter with lambda to keep only non-vowel characters
    result = ''.join(filter(lambda ch: ch not in vowels, text))
    return result

sample = "Shrushti loves Python programming"
print("Original:", sample)
print("Without vowels:", remove_vowels(sample))
```
