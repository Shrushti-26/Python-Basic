```python
#Que6)Write a function to demonstrate the behavior of mutable and immutable arguments.
#Ans:-
def change_number(num): #Function with immutable argument
    print("Inside function (before):", num)
    num= num+10
    print("Inside function (after):", num)
def change_list(lst): #Function with mutable argument
    print("Inside function (before):", lst)
    lst.append(100)
    print("Inside function (after):", lst)

#calling the functions
number=20
my_list =[1,2,3]

print("Before function call (number):", number)
change_number(number)
print("After function call (number):", number)

print("\nBefore function call (list):", my_list)
change_list(my_list)
print("After function call (list):", my_list)
```

```python
#Que14) Write a program to reverse a string using a while loop.
#Ans:-
text=input("Enter a string: ")
rev=""
i=len(text)-1

while i>=0:
    rev=rev+text[i]
    i=i-1
    print("Reversed string:", rev)
```

```python
#Que15) Write a Python program to calculate the factorial of a number provided by the user using a while loop.
#Ans:-
num= int(input("Enter a number "))
fact=1
i=num
whilei>0:
    fact=fact*1
    i-=1
print("Factorial of", num, "is:", fact)
```
