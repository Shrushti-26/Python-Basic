Q.2)What are the key assumptions of Simple Linear Regression?
Ans:-Key Assumptions of Simple Linear Regression:- For Simple Linear Regression to give valid and reliable results, several assumptions must hold true:

1. Linearity
- The relationship between the independent variable (X) and the dependent variable (Y) is linear.
- This means changes in X produce proportional changes in Y.

2. Independence of Errors
- The residuals (errors) are independent of each other.
- No autocorrelation should exist (important in time-series data).

3. Homoscedasticity
- The variance of residuals is constant across all values of X.
- In other words, the spread of errors should be the same at all levels of the predictor.

4. Normality of Errors
- The residuals should be normally distributed.
- This assumption is crucial for hypothesis testing and confidence intervals.

5. No Perfect Multicollinearity (only relevant in multiple regression, but worth noting)
- In simple regression, since there’s only one predictor, this isn’t an issue.
- In multiple regression, predictors should not be highly correlated.

Q.10)What are the key assumptions of Multiple Linear Regression?
Ans:-In Multiple Linear Regression, several assumptions must hold true for the model to be valid and reliable. These assumptions ensure that the estimated coefficients are unbiased and that statistical tests are meaningful.

1. There is the linearity assumption, which means the relationship between each independent variable and the dependent variable is linear. If the true relationship is nonlinear, the model may give misleading results.

2. The independence of errors assumption requires that residuals (the differences between observed and predicted values) are independent of each other. This is especially important in time-series data, where autocorrelation can distort results.

3. The model assumes homoscedasticity, meaning the variance of residuals is constant across all levels of the independent variables. If residuals spread out more at certain values of predictors, the model suffers from heteroscedasticity, which weakens reliability.

4. The normality of errors assumption states that residuals should be normally distributed. This is crucial for hypothesis testing and constructing confidence intervals.

5. A key assumption unique to multiple regression is the absence of multicollinearity. Independent variables should not be highly correlated with each other, because this makes it difficult to isolate the effect of each predictor and can lead to unstable coefficient estimates.

Together, these assumptions form the foundation of Multiple Linear Regression. When they are met, the model provides trustworthy insights into how multiple predictors influence an outcome. If they are violated, diagnostic tests and corrective techniques (like transformations or variable selection) are often needed.

Q.11)What is heteroscedasticity, and how does it affect the results of a Multiple Linear Regression model?
Ans:- Heteroscedasticity occurs when the variance of the residuals (errors) is not constant across all levels of the independent variables. Instead of having a uniform spread, the residuals may fan out or cluster unevenly.

- Effect on Coefficients: The regression coefficients themselves remain unbiased. The line still represents the average relationship between predictors and the outcome.
- Effect on Standard Errors: The standard errors become unreliable. This leads to inaccurate hypothesis tests, confidence intervals, and p-values. You might wrongly conclude that a predictor is significant when it isn’t, or miss a significant predictor.
- Efficiency Loss: Ordinary Least Squares (OLS) estimates are no longer the “best” (minimum variance) estimates. The model loses efficiency, weakening its predictive power.
- Detection: Heteroscedasticity can often be spotted in residual plots, where residuals spread out unevenly. Statistical tests like the Breusch-Pagan test or White’s test can formally check for it.
- Practical Consequence: It undermines the reliability of statistical inference, making conclusions less trustworthy even if the regression line itself is correct.

Q.12)How can you improve a Multiple Linear Regression model with high multicollinearity?
Ans:- some effective ways to improve a Multiple Linear Regression model when you encounter high multicollinearity (strong correlation among independent variables):
- Remove Highly Correlated Predictors: If two or more variables are strongly correlated, consider dropping one of them to reduce redundancy and stabilize coefficient estimates.
- Combine Variables: Sometimes correlated predictors can be combined into a single composite variable (e.g., averaging or creating an index) to capture the shared information without duplication.
- Use Principal Component Analysis (PCA): PCA transforms correlated variables into a smaller set of uncorrelated components. These components can then be used in regression to avoid multicollinearity.
- Apply Regularization Techniques: Methods like Ridge Regression and Lasso Regression add penalties to the regression coefficients. Ridge shrinks coefficients to handle multicollinearity, while Lasso can shrink some coefficients to zero, effectively performing variable selection.
- Centering or Standardizing Variables: Scaling predictors (subtracting the mean or dividing by standard deviation) can sometimes reduce multicollinearity, especially when interaction terms are involved.
- Check Variance Inflation Factor (VIF): Use VIF to identify which variables contribute most to multicollinearity. Variables with very high VIF values are candidates for removal or transformation.

Q.13)What are some common techniques for transforming categorical variables for use in regression models?
Ans:- To use categorical variables in regression models, we need to transform them into numerical formats that the model can interpret. Here are some common techniques:

- Dummy (One-Hot) Encoding:  
Each category is converted into a separate binary variable (0 or 1). For example, a variable "Color" with values Red, Blue, Green becomes three new variables: IsRed, IsBlue, IsGreen. This is the most widely used method.
- Label Encoding:  
Each category is assigned a unique integer. For example, Red = 1, Blue = 2, Green = 3. This is simple but can sometimes mislead the model into thinking there is an ordinal relationship when none exists.
- Ordinal Encoding:  
Used when categories have a natural order (e.g., Small, Medium, Large). Values are mapped to integers that reflect this order, such as Small = 1, Medium = 2, Large = 3.
- Target Encoding (Mean Encoding):  
Categories are replaced with the mean of the dependent variable for that category. For example, if predicting salary, each job title could be encoded with the average salary of people with that title. This can be powerful but risks overfitting.
- Frequency Encoding:  
Categories are replaced with the frequency (count) of their occurrence in the dataset. For example, if "Red" appears 50 times and "Blue" appears 30 times, they are encoded as 50 and 30.
- Binary Encoding:  
Categories are first converted to integers, then represented in binary form, with each digit becoming a separate variable. This reduces dimensionality compared to one-hot encoding when there are many categories.

Q.14)What is the role of interaction terms in Multiple Linear Regression?
Ans:- In Multiple Linear Regression, interaction terms play an important role in capturing relationships that go beyond the simple additive effects of predictors.

An interaction term represents the combined effect of two or more independent variables on the dependent variable. Instead of assuming that each predictor influences the outcome independently, interaction terms allow us to model situations where the effect of one variable depends on the level of another.

For example, suppose you are predicting salary based on years of experience (X₁) and education level (X₂). Without interaction, the model assumes that the effect of experience is the same regardless of education. But in reality, the impact of experience might be stronger for people with higher education. By including an interaction term (𝑋1⋅𝑋2), the model can capture this combined influence.

The role of interaction terms can be summarized as:
- They reveal conditional relationships, showing how predictors work together rather than in isolation.
- They increase model flexibility, allowing regression to represent more complex real-world patterns.
- They help avoid oversimplification, ensuring that important combined effects are not missed.

In short, interaction terms enrich a regression model by accounting for situations where the effect of one variable is not constant but depends on another variable.

Q.15)How can the interpretation of intercept differ between Simple and Multiple Linear Regression?
Ans:- In Simple Linear Regression, the intercept represents the expected value of the dependent variable when the single independent variable equals zero. For example, if you are predicting exam scores based on hours studied, the intercept would be the predicted score for a student who studied zero hours. It’s straightforward because there is only one predictor, so the intercept is simply the baseline outcome when that predictor is absent.

In Multiple Linear Regression, the interpretation of the intercept becomes more nuanced. It represents the expected value of the dependent variable when all independent variables equal zero simultaneously. This can sometimes be meaningful—for instance, predicting salary when experience, education, and weekly hours worked are all zero—but often it is not realistic or interpretable, especially if zero is not a valid or meaningful value for some predictors (like education level or age). In such cases, the intercept is still mathematically necessary for positioning the regression plane, but its practical interpretation may be limited.

So, the key difference is that in simple regression the intercept is the baseline outcome with one predictor at zero, while in multiple regression it is the baseline outcome with all predictors at zero, which may or may not make practical sense depending on the variables involved.

Q.18)What are the limitations of using R² as a sole measure of model performance?
Ans:- Using R² alone to judge the performance of a regression model can be misleading because it doesn’t capture the full picture. Here are the main limitations:

- Doesn’t Indicate Causality:  
A high R² only shows that the model explains a large portion of the variance in the dependent variable. It doesn’t prove that the predictors cause the changes in the outcome.

- Insensitive to Overfitting:  
Adding more predictors will almost always increase R², even if those predictors are irrelevant. This can make the model look better than it really is.

- No Insight into Bias or Model Validity:  
R² doesn’t tell you whether the regression assumptions (linearity, independence, homoscedasticity, normality) are satisfied. A model can have a high R² but still be invalid.

- Doesn’t Reflect Predictive Accuracy:  
R² measures how well the model fits the training data, but it doesn’t guarantee good performance on new, unseen data. Cross-validation or test set performance is needed for that.

- Can Be Misleading in Nonlinear Relationships:  
If the true relationship between variables is nonlinear, R² from a linear model may underestimate or misrepresent the fit.

- Low R² Doesn’t Always Mean Useless Model:  
In fields like social sciences, even a low R² can still provide meaningful insights if the predictors are statistically significant.

In short, R² is a useful measure of goodness-of-fit, but it should be complemented with other metrics like Adjusted R², RMSE, MAE, cross-validation scores, and residual analysis to properly evaluate model performance.

Q.19)How would you interpret a large standard error for a regression coefficient?
Ans:-A large standard error for a regression coefficient signals that the estimate of that coefficient is highly uncertain. Here’s how to interpret it:

- Unstable Estimate:  
The coefficient value is not precise. The model is struggling to pin down the true effect of that predictor on the dependent variable.

- Possible Multicollinearity:  
Large standard errors often arise when predictors are highly correlated with each other. This makes it difficult for the model to separate their individual contributions.

- Weak Predictor Influence:  
It may indicate that the predictor doesn’t have a strong or consistent relationship with the outcome, so its effect varies widely across the data.

- Impact on Hypothesis Testing:  
A large standard error increases the likelihood that the t-statistic for the coefficient will be small, leading to a high p-value. This means the predictor may not be statistically significant.

- Practical Consequence:  
Predictions involving that variable become less reliable, since the model cannot confidently estimate how changes in the predictor affect the outcome.

In short, a large standard error suggests that the coefficient is unstable and possibly unreliable, often due to multicollinearity or weak explanatory power. Analysts typically check Variance Inflation Factor (VIF) or consider removing/restructuring predictors to address this issue.

Q.20)How can heteroscedasticity be identified in residual plots, and why is it important to address it?
Ans:- Heteroscedasticity can be identified in residual plots by looking at how the residuals (errors) are spread across the fitted values or predictors:

1. Visual Signs in Residual Plots:
- If residuals are evenly scattered around zero with roughly equal spread, the variance is constant (homoscedasticity).
- If residuals show a funnel shape (narrow at one end, wider at the other), or a pattern where variance increases or decreases systematically, this indicates heteroscedasticity.
- Clustering or uneven “bands” of residuals also suggest non-constant variance.

2. Why It’s Important to Address:

- Unreliable Standard Errors: Heteroscedasticity makes the estimated standard errors incorrect, which can lead to misleading t-tests, p-values, and confidence intervals.
- Weak Statistical Inference: You may wrongly conclude that predictors are significant (or insignificant).
- Loss of Efficiency: Ordinary Least Squares (OLS) estimates are no longer the “best” (minimum variance) estimates, reducing the reliability of predictions.
- Model Diagnostics: Ignoring heteroscedasticity can mask deeper issues in the data or model specification.

In short, residual plots are a diagnostic tool: they visually reveal heteroscedasticity, and addressing it is crucial to ensure that regression results are statistically valid and trustworthy.

Q.21)What does it mean if a Multiple Linear Regression model has a high R² but low adjusted R²?
Ans:- If a Multiple Linear Regression model shows a high R² but a low Adjusted R², it usually means the model is being inflated by unnecessary predictors.

- High R²: This indicates that the model explains a large proportion of the variance in the dependent variable. However, R² always increases (or at least stays the same) when more predictors are added, even if those predictors don’t truly contribute to explaining the outcome.

- Low Adjusted R²: Adjusted R² penalizes the model for including predictors that don’t add meaningful explanatory power. If it drops significantly compared to R², it suggests that many of the predictors are irrelevant or redundant, and the model is overfitting the data.

- Interpretation: In this situation, the model may look impressive at first glance (high R²), but the low Adjusted R² reveals that the apparent fit is misleading. The model is likely cluttered with predictors that don’t improve its explanatory strength and may even reduce generalizability to new data.

- Practical Implication: You should reconsider the variables included in the model. Techniques like checking p-values, Variance Inflation Factor (VIF), or stepwise regression can help identify and remove unnecessary predictors, leading to a more parsimonious and reliable model.

In short, high R² with low Adjusted R² is a red flag for overfitting—the model is explaining variance artificially by adding predictors that don’t truly matter.

Q.22)Why is it important to scale variables in Multiple Linear Regression?
Ans:- Scaling variables in Multiple Linear Regression is important because it ensures that the model treats all predictors fairly and avoids distortions caused by differences in measurement units or ranges. Here’s why it matters:

- Equal Contribution of Predictors:  
When variables are measured on very different scales (e.g., income in thousands vs. age in years), the regression coefficients can be disproportionately influenced by the larger-scaled variable. Scaling puts them on a comparable footing.

- Improved Numerical Stability:  
Regression involves matrix operations. If variables have vastly different magnitudes, calculations can become unstable, leading to inaccurate or unreliable coefficient estimates. Scaling reduces this risk.

- Interpretability of Coefficients:  
After scaling, coefficients reflect the effect of a one standard deviation change in the predictor, making comparisons across variables more meaningful.

- Better Detection of Multicollinearity:  
Scaling helps in diagnosing multicollinearity (using tools like Variance Inflation Factor), because it removes distortions caused by scale differences.

- Essential for Regularization Methods:  
Techniques like Ridge Regression and Lasso Regression penalize coefficients. Without scaling, variables with larger ranges can dominate the penalty, skewing results. Scaling ensures penalties are applied fairly.

In short, scaling doesn’t change the underlying relationships but makes the regression model more stable, interpretable, and fair across predictors.

Q.23)What is polynomial regression?
Ans:- Polynomial Regression is an extension of linear regression that allows us to model nonlinear relationships between the independent variable(s) and the dependent variable by including higher-order powers of the predictors.

Polynomial regression fits a regression equation of the form:
𝑌=𝛽0+𝛽1𝑋+𝛽2𝑋2+𝛽3𝑋3+⋯+𝛽𝑛𝑋𝑛+𝜖

- Flexibility:  
By increasing the degree of the polynomial, the model can fit more complex patterns. However, too high a degree can lead to overfitting.

- Interpretation:  
Unlike linear regression, the coefficients in polynomial regression are harder to interpret directly. Instead, the focus is on how well the curve fits the data.

- Applications:  
Commonly used in situations where trends are nonlinear, such as growth curves, economics, or physics experiments.

- Limitations:
    - Risk of overfitting with high-degree polynomials.
    - Extrapolation beyond the data range can be highly unreliable.
    - Multicollinearity can arise because polynomial terms are correlated with each other.

Q.28)What are the limitations of polynomial regression?
Ans:- Polynomial regression is powerful for modeling nonlinear relationships, but it comes with several limitations that are important to keep in mind:

1. Risk of Overfitting:  
Higher-degree polynomials can fit the training data extremely well, but they may capture noise rather than the true underlying pattern, leading to poor generalization on new data.

2. Unstable Predictions Outside Data Range:  
Polynomial curves can swing wildly when extrapolated beyond the observed data, making predictions unreliable.

3. Interpretability Issues:  
Coefficients of higher-order terms are difficult to interpret in a meaningful way, unlike linear regression where slopes have clear practical meaning.

4. Multicollinearity:  
Polynomial terms (e.g., 𝑋, 𝑋2, 𝑋3) are often highly correlated with each other, which inflates standard errors and makes coefficient estimates unstable.

5. Computational Complexity:  
As the number of predictors and polynomial degree increases, the number of terms grows rapidly, making the model complex and harder to manage.

6. Bias-Variance Tradeoff:  
Low-degree polynomials may underfit (too simple), while high-degree polynomials may overfit (too complex). Choosing the right degree is tricky and often requires cross-validation.

In short: Polynomial regression is useful for capturing nonlinear trends, but it must be applied carefully. Analysts often use regularization methods (Ridge, Lasso) or splines as alternatives to balance flexibility with stability.

Q.29)What methods can be used to evaluate model fit when selecting the degree of a polynomial?
Ans:- When selecting the degree of a polynomial in regression, you want to balance flexibility (capturing nonlinear patterns) with simplicity (avoiding overfitting). Several methods can help evaluate model fit:

Key Evaluation Methods:-
1. Residual Plots:  
Check if residuals are randomly scattered. If systematic patterns remain, a higher-degree polynomial may be needed. If residuals look random, the chosen degree is adequate.

2. R² vs. Adjusted R²:
- R² always increases with more terms.
- Adjusted R² penalizes unnecessary complexity. A drop in Adjusted R² suggests the added polynomial terms don’t improve the model meaningfully.

3. Cross-Validation (CV):  
Use techniques like k-fold CV to compare prediction errors across polynomial degrees. The degree with the lowest average error is usually best.

4. Information Criteria (AIC/BIC):  
These metrics balance model fit with complexity. Lower AIC/BIC values indicate a better trade-off between accuracy and simplicity.

5. Mean Squared Error (MSE) / Root Mean Squared Error (RMSE):  
Evaluate on a validation or test set. A lower error indicates better predictive performance.

6. F-Test for Nested Models:  
Compare models of different degrees to see if adding higher-order terms significantly improves fit.

Q.30)Why is visualization important in polynomial regression?
Ans:- Visualization is especially important in polynomial regression because it helps you understand how well the model is capturing nonlinear patterns and whether it’s behaving reasonably.

Why Visualization Matters:- 
1. Detecting Nonlinear Fit:  
Polynomial regression bends the line to fit curves. Visualizing the fitted curve against the data points shows whether the chosen degree captures the true trend or is under/overfitting.

2. Spotting Overfitting:  
Higher-degree polynomials can produce wavy, erratic curves that fit the training data too closely. A plot makes this obvious, whereas metrics alone might hide it.

3. Residual Analysis:  
Plotting residuals helps check if errors are randomly distributed. If residuals show patterns, the polynomial degree may not be appropriate.

4. Interpretability:  
Polynomial coefficients are hard to interpret directly. Visualization translates them into a curve that’s easier to understand and communicate.

5. Extrapolation Risks:  
Polynomial models can behave unpredictably outside the observed data range. Visualizing the curve beyond the data points highlights these risks.

```python
# Q.31)How is polynomial regression implemented in Python?
# Ans:-

import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
# Example dataset
X = np.array([1, 2, 3, 4, 5, 6]).reshape(-1, 1)  # predictor
y = np.array([2.3, 2.9, 4.1, 5.8, 8.0, 10.1])    # target
```

```python
# Transform Predictors into Polynomial Features
poly = PolynomialFeatures(degree=2)   # choose polynomial degree
X_poly = poly.fit_transform(X)        # adds X^2 term

#Fit Linear Regression on Polynomial Features
model = LinearRegression()
model.fit(X_poly, y)

#Make Predictions
y_pred = model.predict(X_poly)

#Visualize Fit
plt.scatter(X, y, color='blue', label='Data')
plt.plot(X, y_pred, color='red', label='Polynomial Fit')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()
```
