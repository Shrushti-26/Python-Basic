Que.1) What is a random variable in probability theory? 
Ans:-
A random variable is a fundamental concept in probability and statistics. It provides a way to assign numerical values to the outcomes of a random experiment.
Definition-> A random variable is a function that maps outcomes from a sample space to real numbers.
Notation: Random variables are usually denoted by capital letters like X,Y,Z.
Purpose: They connect theoretical probability with real-world data, allowing us to use statistical tools.
Types:
- Discrete random variable: Takes on countable values (e.g., number of heads in coin tosses).
- Continuous random variable: Takes on values from an interval of real numbers (e.g., height, time).
Example
1) Rolling a die: The sample space is { 1,2,3,4,5,6} .
Define random variable X as "the number showing on the die."
Then X maps each outcome directly to its numeric value.
2) Flipping two coins: The sample space is { HH,HT,TH,TT} .
Define random variable Y as "the number of heads."
Then Y(HH)=2, Y(HT)=1, Y(TH)=1, Y(TT)=0.
Random variables are the foundation for Probability distributions (e.g., binomial, normal),Expected value and variance, Statistical inference and modeling.

Que.4)What is a binomial distribution, and how is it used in probability? 
Ans:-
The Binomial Distribution is one of the most important discrete probability distributions in statistics and probability theory.
It models the number of successes in a fixed number of independent trials, where:
- Each trial has only two possible outcomes: success or failure.
- The probability of success (p) is the same for each trial.
- The number of trials is fixed (n).
Formally, if X is a binomial random variable:

where:
n = number of trials
p = probability of success
k = number of successes
{n \choose k} = number of ways to choose k successes out of n trials
  
Properties
- Mean: E[X]=np
- Variance: \mathrm{Var}(X)=np(1-p)
- Distribution type: Discrete

Examples
- Tossing a coin 10 times and counting the number of heads.
- Quality control: Number of defective items in a batch of 20.
- Survey predictions: Number of people who say "yes" in a sample of 100 respondents.

Que.5) What is the standard normal distribution, and why is it important? 
Ans:-
The Standard Normal Distribution is a special case of the normal (Gaussian) distribution that plays a foundational role in probability and statistics.
It is a normal distribution with:
- Mean (\mu ) = 0
- Standard deviation (\sigma ) = 1
The curve is symmetric around 0 and shaped like the familiar bell curve.
Applications :- Hypothesis testing, confidence intervals, data standardization

In short, the standard normal distribution is the benchmark that makes statistical analysis possible. By converting data into Z-scores, we can apply universal probability tables and compare results across different contexts.

Que.6) What is the Central Limit Theorem (CLT), and why is it critical in statistics? 
Ans:-
The Central Limit Theorem (CLT) is one of the most fundamental results in probability and statistics. It explains why the normal distribution appears so frequently in data analysis, even when the underlying population distribution is not normal.
- If you take sufficiently large samples (n>30 is a common rule of thumb) from any population with a finite mean and variance,
- The distribution of the sample mean (or sum) will tend to follow a normal distribution, regardless of the population’s original shape (skewed, uniform, etc.).

The CLT is the bridge between raw data and statistical inference. It allows us to use the powerful tools of the normal distribution even when the population itself is unknown or irregular.

Que.7) What is the significance of confidence intervals in statistical analysis? ?
Ans:-
Confidence intervals are a cornerstone of statistical analysis because they provide a range of plausible values for a population parameter (like a mean or proportion), rather than just a single estimate.
A confidence interval (CI) gives an estimated range of values that is likely to include the true population parameter.
Confidence intervals combine statistical accuracy with uncertainty measurement, making them essential for interpreting results responsibly. They prevent overconfidence in point estimates and help researchers communicate the reliability of findings

*Confidence Intervals Are Significant
- Measure of Precision - They show how reliable an estimate is. Narrow intervals mean high precision; wide intervals mean more uncertainty.
- Better than Point Estimates - A single number (like a sample mean) doesn’t capture uncertainty. Confidence intervals provide context.
- Linked to Significance Testing - If a CI for a difference between groups does not include 0, it suggests a statistically significant difference.
- Decision-Making Tool - Used in medicine, business, and social sciences to assess risk, effectiveness, or reliability of results.

Que.8) What is the concept of expected value in a probability distribution? 
Ans:-
The expected value (also called the mean of a random variable) is a fundamental concept in probability and statistics. It represents the long-run average outcome of a random variable if an experiment were repeated many times.

**Intuition
- Think of expected value as the weighted average of all possible outcomes, weighted by their probabilities.
- It’s not necessarily a value that the random variable can actually take, but rather the center of gravity of the distribution.

**Applications
- Decision-making: Used in economics, finance, and gambling to evaluate expected gains/losses.
- Statistics: Forms the basis of measures like variance and standard deviation.
- Predictive modeling: Helps forecast outcomes in uncertain scenarios.



```python
#Que.9)  Write a Python program to generate 1000 random numbers from a normal distribution with mean = 50 and standard deviation = 5. Compute its mean and standard deviation using NumPy, and draw a histogram to visualize the distribution. 
#Ans:-
import numpy as np
import matplotlib.pyplot as plt

# Generate 1000 random numbers from a normal distribution
data = np.random.normal(loc=50, scale=5, size=1000)

# Compute mean and standard deviation
mean = np.mean(data)
std_dev = np.std(data)

print(f"Mean: {mean:.2f}")
print(f"Standard Deviation: {std_dev:.2f}")

# Plot histogram
plt.hist(data, bins=30, color='skyblue', edgecolor='black')
plt.title("Histogram of Normal Distribution (mean=50, std=5)")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.axvline(mean, color='red', linestyle='dashed', linewidth=2, label=f"Mean = {mean:.2f}")
plt.legend()
plt.show()
```

```python
#Que.10)You are working as a data analyst for a retail company. The company has collected daily sales data for 2 years and wants you to identify the overall sales trend. 
#daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255, 235, 260, 245, 250, 225, 270, 265, 255, 250, 260] 
#● Explain how you would apply the Central Limit Theorem to estimate the average sales with a 95% confidence interval. 
#● Write the Python code to compute the mean sales and its confidence interval. 

#Ans:-
import numpy as np
import scipy.stats as st

# Daily sales data (sample provided)
daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255,
               235, 260, 245, 250, 225, 270, 265, 255, 250, 260]

# Convert to NumPy array
data = np.array(daily_sales)

# Compute mean and standard deviation
mean_sales = np.mean(data)
std_sales = np.std(data, ddof=1)  # sample standard deviation
n = len(data)

# 95% confidence interval using CLT
confidence_level = 0.95
alpha = 1 - confidence_level
z_score = st.norm.ppf(1 - alpha/2)  # 1.96 for 95% CI

margin_of_error = z_score * (std_sales / np.sqrt(n))
ci_lower = mean_sales - margin_of_error
ci_upper = mean_sales + margin_of_error

print(f"Mean Sales: {mean_sales:.2f}")
print(f"95% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")
```
