# Que.1) What is the difference between descriptive statistics and inferential statistics? Explain with examples. 
# Ans:- 
Descriptive Statistics
- Definition: Descriptive statistics summarize and organize data so it can be easily understood. They describe what has happened in a dataset without making predictions or generalizations.
- Purpose: To present raw data in a meaningful way.
- Techniques: Measures of central tendency (mean, median, mode), measures of dispersion (range, variance, standard deviation), frequency distributions, and data visualization (charts, graphs).
- Example:
Suppose a teacher records the marks of 30 students in a math test.
- The average score is 75.
- The highest score is 95, and the lowest is 50.
- A histogram shows how scores are distributed.
These statistics describe the dataset but do not go beyond it.

Inferential Statistics
- Definition: Inferential statistics use data from a sample to make conclusions, predictions, or generalizations about a larger population.
- Purpose: To infer patterns and test hypotheses beyond the immediate dataset.
- Techniques: Hypothesis testing, confidence intervals, regression analysis, correlation, ANOVA, chi-square tests.
- Example:
A researcher wants to know the average height of all college students in India. Measuring every student is impossible, so they collect a sample of 500 students.
- The sample mean height is 168 cm.
- Using inferential statistics, they estimate that the population mean height lies between 166–170 cm with 95% confidence.
- They might also test whether male students are significantly taller than female students.
These conclusions go beyond the sample and apply to the population.


# Que.2)  What is sampling in statistics? Explain the differences between random and stratified sampling. 
# Ans:-
Sampling in statistics is the process of selecting a smaller group (called a sample) from a larger population so that studying the sample can provide insights about the whole population. Since it’s often impractical to collect data from every individual, sampling makes research more efficient while still allowing reliable conclusions.

Random sampling means that every individual in the population has an equal chance of being chosen. The selection is purely by chance, like drawing names from a hat or using a random number generator. This method is simple and unbiased, but if the population is very diverse, it might not capture all subgroups well. For example, if a researcher randomly selects 200 students from a university, each student has the same probability of being included, but some faculties might end up underrepresented.

Stratified sampling, on the other hand, involves dividing the population into subgroups (called strata) based on certain characteristics such as age, gender, or faculty. Then, samples are taken from each subgroup in proportion to their size in the population. This ensures that all important subgroups are represented. For instance, if the same researcher wants to study eating habits, they might divide students by faculty (Engineering, Arts, Science, Commerce) and then randomly select students from each faculty according to its proportion in the university.

In short, random sampling relies on chance and treats everyone equally, while stratified sampling deliberately ensures that all key subgroups are included in the sample.

# Que.3) Define mean, median, and mode. Explain why these measures of central tendency are important. 
# Ans:-
Mean
The mean (often called the average) is calculated by adding up all the values in a dataset and dividing by the number of values.
- Example: For the numbers 2, 4, 6, 8, the mean is (2+4+6+8)/4=5.
- It gives a single value that represents the "typical" magnitude of the data.

Median
The median is the middle value when the data is arranged in order. If there is an even number of values, the median is the average of the two middle values.
- Example: For the numbers 2, 4, 6, 8, the median is (4+6)/2=5.
- It is useful because it is not affected by extreme values (outliers).

Mode
The mode is the value that occurs most frequently in a dataset. A dataset can have one mode (unimodal), more than one mode (bimodal or multimodal), or no mode if all values occur with equal frequency.
- Example: In the dataset 2, 4, 4, 6, 8, the mode is 4.
- It highlights the most common or popular value.

In essence, mean, median, and mode are complementary tools: the mean shows the overall average, the median shows the middle ground, and the mode shows the most frequent occurrence. Together, they provide a fuller picture of the dataset.

# Que.4)Explain skewness and kurtosis. What does a positive skew imply about the data? 
# Ans:-
Skewness
Skewness measures the asymmetry of a data distribution around its mean.
Types:-
- Zero skewness → perfectly symmetrical distribution (like a normal distribution).
- Positive skewness → the right tail of the distribution is longer; most values are concentrated on the left, but a few very large values pull the mean to the right.
- Negative skewness → the left tail is longer; most values are concentrated on the right, but a few very small values pull the mean to the left.

Kurtosis
Kurtosis measures the tailedness or the sharpness of the peak of a distribution compared to a normal distribution.
Types:-
- Mesokurtic (kurtosis ≈ 3) → similar to a normal distribution.
- Leptokurtic (kurtosis > 3) → distribution has heavy tails and a sharp peak; more prone to outliers.
- Platykurtic (kurtosis < 3) → distribution has light tails and a flatter peak; fewer extreme values.

```python
#Que.6) Compute the covariance and correlation coefficient between the following two datasets provided as lists in Python: 
list_x = [10, 20, 30, 40, 50] 
list_y = [15, 25, 35, 45, 60] 

import numpy as np

# Convert to numpy arrays
x = np.array(list_x)
y = np.array(list_y)

# Compute covariance matrix
cov_matrix = np.cov(x, y, bias=True)   # bias=True → population covariance
cov_xy = cov_matrix[0, 1]

# Compute correlation coefficient
corr_xy = np.corrcoef(x, y)[0, 1]

print("Covariance:", cov_xy)
print("Correlation Coefficient:", corr_xy)
```

```python
#Que.7)  Write a Python script to draw a boxplot for the following numeric list and identify its outliers. Explain the result: 
#Ans:-
import matplotlib.pyplot as plt
import numpy as np

data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]

# Create boxplot
plt.boxplot(data)
plt.title("Boxplot of Data")
plt.ylabel("Values")
plt.show()

# Calculate quartiles
Q1 = np.percentile(data, 25)
Q3 = np.percentile(data, 75)
IQR = Q3 - Q1

# Define outlier thresholds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers
outliers = [x for x in data if x < lower_bound or x > upper_bound]

print("Q1:", Q1)
print("Q3:", Q3)
print("IQR:", IQR)
print("Lower Bound:", lower_bound)
print("Upper Bound:", upper_bound)
print("Outliers:", outliers)
```
