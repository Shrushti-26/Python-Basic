#  Q.1)  What is hypothesis testing in statistics? 
Ans :- Hypothesis testing is a statistical method used to make decisions or draw conclusions about a population based on sample data. It helps determine whether there is enough evidence to support a particular claim or assumption.

Types of Hypothesis:
1) Null Hypothesis (H‚ÇÄ):  
The default assumption, often stating "no effect" or "no difference."
Example: The average website visits per day = 50.

2) Alternative Hypothesis (H‚ÇÅ):  
The competing claim suggesting there is an effect or difference.
Example: The average website visits per day ‚â† 50.

3) Significance Level (Œ±):  
The threshold for deciding whether to reject H‚ÇÄ, commonly set at 0.05 (5%). It represents the probability of rejecting H‚ÇÄ when it is actually true.

4) Test Statistic:  
A value calculated from sample data (e.g., z-score, t-score) used to compare against a critical value.

5) p-value:  
The probability of observing the sample results if H‚ÇÄ is true.
If p-value ‚â§ Œ±, reject H‚ÇÄ (evidence supports H‚ÇÅ).
If p-value > Œ±, fail to reject H‚ÇÄ (insufficient evidence against H‚ÇÄ).

Steps in Hypothesis Testing:
1)State the hypotheses (H‚ÇÄ and H‚ÇÅ).
2)Choose significance level (Œ±).
3)Collect and analyze sample data.
4)Compute test statistic and p-value.
5)Make a decision: reject or fail to reject H‚ÇÄ.
6)Draw a conclusion in context of the problem.

# Q.3)Explain the significance level in hypothesis testing and its role in deciding the outcome of a test. 
Ans:-The significance level (Œ±) is a threshold used in hypothesis testing to decide whether to reject the null hypothesis (H‚ÇÄ). It represents the probability of making a Type I error‚Äîrejecting H‚ÇÄ when it is actually true.
Definition: Œ± is the maximum risk you are willing to take of incorrectly rejecting H‚ÇÄ.
Common Values: 0.05 (5%), 0.01 (1%), or 0.10 (10%).
Interpretation:
If the p-value ‚â§ Œ±, reject H‚ÇÄ (evidence supports H‚ÇÅ).
If the p-value > Œ±, fail to reject H‚ÇÄ (insufficient evidence against H‚ÇÄ).

Role in Decision-Making:
1)Controls error rate: By setting Œ±, researchers limit the chance of falsely claiming an effect exists.
2)Balances risk: A smaller Œ± (like 0.01) reduces false positives but increases the chance of missing a real effect (Type II error).
3)Context-dependent:Medical trials often use Œ± = 0.01 (to minimize false claims of effectiveness), Social sciences often use Œ± = 0.05 (a balance between sensitivity and caution).

# Q.4)What are Type I and Type II errors? Give examples of each. 
Ans:- When we test hypotheses, two kinds of mistakes can occur depending on how we interpret the sample evidence. These are called Type I and Type II errors.

Type I Error (False Positive):Type I Error rejecting the null hypothesis (H‚ÇÄ) when it is actually true.
Cause: Evidence appears strong enough to suggest an effect, but in reality, there is none.
Probability: Equal to the significance level (Œ±).
Example:
Medical trial: Concluding a new drug works when it actually has no effect.
Courtroom analogy: Convicting an innocent person.

Type II Error (False Negative):Typr II Error failing to reject the null hypothesis when the alternative hypothesis (H‚ÇÅ) is actually true.
Cause: The test lacks enough power to detect the real effect.
Probability: Denoted by Œ≤ (beta).
Example:
Medical trial: Concluding a new drug has no effect when it actually does help patients.
Courtroom analogy: Letting a guilty person go free.

# Q.5)What is the difference between a Z-test and a T-test? Explain when to use each.
Ans:- Both Z-tests and T-tests are statistical methods used in hypothesis testing to compare means, but they differ in assumptions about the population and sample size.

Z-test: A test based on the standard normal distribution (Z-distribution).
When to Use: 1) Large sample size (typically ùëõ‚â•30).
2) Population variance (œÉ¬≤) or standard deviation (œÉ) is known.

Applications:
- Comparing a sample mean to a population mean.
- Comparing means of two large samples.

Example: Testing whether the average height of 1,000 students differs from the national average when the population standard deviation is known.

T-test: A test based on the Student‚Äôs t-distribution.
When to Use: 1) Small sample size (ùëõ<30).
2)Population variance is unknown (estimated from the sample).

Applications:
- One-sample t-test: Compare a sample mean to a hypothesized population mean.
- Independent two-sample t-test: Compare means of two independent groups.
- Paired t-test: Compare means of two related samples (e.g., before-and-after measurements).

Example: Testing whether the average exam score of 20 students differs from 70 when the population variance is unknown.

```python
# Q.7)Implement hypothesis testing using Z-statistics for a sample dataset in Python. Show the Python code and interpret the results.
#Ans:-
import numpy as np
from scipy.stats import norm

# Sample data
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

# Hypothesized population mean
mu0 = 50

# Known population standard deviation (assumed for Z-test)
sigma = 0.5  

# Sample statistics
sample_mean = np.mean(sample_data)
n = len(sample_data)

# Z-statistic
z_stat = (sample_mean - mu0) / (sigma / np.sqrt(n))

# Two-tailed p-value
p_value = 2 * (1 - norm.cdf(abs(z_stat)))

print("Sample Mean:", sample_mean)
print("Z-statistic:", z_stat)
print("p-value:", p_value)

# Decision at alpha = 0.05
alpha = 0.05
if p_value < alpha:
    print("Reject H0: The sample mean is significantly different from", mu0)
else:
    print("Fail to reject H0: No significant difference from", mu0)
```

```python
# Q.8)  Write a Python script to simulate data from a normal distribution and calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.
# Ans:-
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Step 1: Simulate data from a normal distribution
np.random.seed(42)  # for reproducibility
mu, sigma = 50, 5   # mean and standard deviation
n = 100             # sample size
data = np.random.normal(mu, sigma, n)

# Step 2: Calculate sample mean and standard error
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)  # sample standard deviation
se = sample_std / np.sqrt(n)

# Step 3: Compute 95% confidence interval
confidence_level = 0.95
z_score = stats.norm.ppf(1 - (1 - confidence_level)/2)  # 1.96 for 95%
margin_of_error = z_score * se
ci_lower = sample_mean - margin_of_error
ci_upper = sample_mean + margin_of_error

print("Sample Mean:", sample_mean)
print("95% Confidence Interval: ({:.2f}, {:.2f})".format(ci_lower, ci_upper))

# Step 4: Plot the data
plt.hist(data, bins=15, edgecolor='black', alpha=0.7)
plt.axvline(sample_mean, color='red', linestyle='dashed', linewidth=2, label=f"Mean = {sample_mean:.2f}")
plt.axvline(ci_lower, color='green', linestyle='dashed', linewidth=2, label=f"Lower CI = {ci_lower:.2f}")
plt.axvline(ci_upper, color='green', linestyle='dashed', linewidth=2, label=f"Upper CI = {ci_upper:.2f}")
plt.title("Normal Distribution Simulation with 95% CI")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```

```python
# Q.9)  Write a Python function to calculate the Z-scores from a dataset and visualize the standardized data using a histogram. Explain what the Z-scores represent in terms of standard deviations from the mean.
# Ans:-
import numpy as np
import matplotlib.pyplot as plt

def z_score_standardization(data):
    """
    Calculate Z-scores for a dataset and plot histogram.
    """
    mean = np.mean(data)
    std = np.std(data, ddof=1)  # sample standard deviation
    
    # Calculate Z-scores
    z_scores = (data - mean) / std
    
    # Print summary
    print("Mean of data:", mean)
    print("Standard deviation of data:", std)
    print("First 5 Z-scores:", z_scores[:5])
    
    # Plot histogram of Z-scores
    plt.hist(z_scores, bins=15, edgecolor='black', alpha=0.7)
    plt.title("Histogram of Z-scores (Standardized Data)")
    plt.xlabel("Z-score")
    plt.ylabel("Frequency")
    plt.axvline(0, color='red', linestyle='dashed', linewidth=2, label="Mean (Z=0)")
    plt.legend()
    plt.show()
    
    return z_scores

# Example usage
data = np.random.normal(50, 5, 100)  # simulate normal distribution data
z_scores = z_score_standardization(data)
```
