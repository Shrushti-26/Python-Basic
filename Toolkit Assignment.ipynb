```python
#Que.2)Write a Pandas script to find the mean of a specific column in a DataFrame
#Ans:-
import pandas as pd
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [25, 30, 35, 40],
    'Salary': [50000, 60000, 70000, 80000]
}

df = pd.DataFrame(data)
# Find the mean of a specific column (e.g., 'Salary')
mean_salary = df['Salary'].mean()

print("Mean of Salary column:", mean_salary)
```

```python
#Que.3)Create a scatter plot using Matplotlib
#Ans:-
import matplotlib.pyplot as plt
x = [5, 7, 8, 7, 2, 17, 2, 9, 4, 11]
y = [99, 86, 87, 88, 100, 86, 103, 87, 94, 78]

# Create scatter plot
plt.scatter(x, y, color='blue', marker='o', s=100)
plt.xlabel("X-axis values")
plt.ylabel("Y-axis values")
plt.title("Simple Scatter Plot")
plt.show()
```

```python
#Que.4)How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap?
#Ans:-
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = {
    'Age': [25, 30, 35, 40, 45],
    'Salary': [50000, 60000, 70000, 80000, 90000],
    'Experience': [2, 5, 7, 10, 12]
}

df = pd.DataFrame(data)

# Calculate correlation matrix
corr_matrix = df.corr()

# Plot heatmap
plt.figure(figsize=(6,4))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)

plt.title("Correlation Matrix Heatmap")
plt.show()
```

```python
#Que.5)Generate a bar plot using Plotly
#Ans:-
import plotly.express as px
import pandas as pd

data = {
    'Fruit': ['Apple', 'Banana', 'Orange', 'Grapes'],
    'Quantity': [10, 15, 7, 20]
}

df = pd.DataFrame(data)

# Create bar plot
fig = px.bar(df, x='Fruit', y='Quantity', color='Fruit',
             title="Bar Plot of Fruit Quantities")
fig.show()
```

```python
#Que.7)Write a program to perform element-wise multiplication of two NumPy arrays.
#Ans:-
import numpy as np

arr1 = np.array([1, 2, 3, 4])
arr2 = np.array([10, 20, 30, 40])

result = arr1 * arr2

print("Array 1:", arr1)
print("Array 2:", arr2)
print("Element-wise multiplication result:", result)
```

```python
#Que.8)Create a line plot with multiple lines using Matplotlib
#Ans:-
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]

y1 = [2, 4, 6, 8, 10]   # Line 1
y2 = [1, 3, 5, 7, 9]    # Line 2
y3 = [2, 3, 4, 5, 6]    # Line 3

# Plot multiple lines
plt.plot(x, y1, label='Line 1', color='blue', marker='o')
plt.plot(x, y2, label='Line 2', color='red', linestyle='--', marker='s')
plt.plot(x, y3, label='Line 3', color='green', linestyle='-.', marker='^')

# Add labels and title
plt.xlabel("X-axis values")
plt.ylabel("Y-axis values")
plt.title("Line Plot with Multiple Lines")

plt.legend()
plt.show()
```

```python
#Que.9)Generate a Pandas DataFrame and filter rows where a column value is greater than a threshold
#Ans:-
import pandas as pd

data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Age': [25, 30, 35, 40, 22],
    'Salary': [50000, 60000, 70000, 80000, 45000]
}
df = pd.DataFrame(data)

threshold = 60000
filtered_df = df[df['Salary'] > threshold]

print("Original DataFrame:\n", df)
print("\nFiltered DataFrame (Salary > 60000):\n", filtered_df)
```

```python
#Que.10)Create a histogram using Seaborn to visualize a distribution
#Ans:-
import seaborn as sns
import matplotlib.pyplot as plt

data = [12, 15, 14, 10, 18, 20, 22, 19, 17, 16, 14, 13, 15, 21, 23, 25, 19, 18, 20]

# Create histogram
sns.histplot(data, bins=8, kde=True, color='skyblue')

plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Histogram of Sample Data")
plt.show()
```

```python
#Que.11)Perform matrix multiplication using NumPy
#Ans:-
import numpy as np

# Define two matrices (2D arrays)
A = np.array([[1, 2],
              [3, 4]])

B = np.array([[5, 6],
              [7, 8]])

result =np.dot(A, B) # or A @ B

print("Matrix A:\n", A)
print("Matrix B:\n", B)
print("Matrix Multiplication Result:\n", result)
```
